```python
"""Integration in Python (8 pts)
We will be using different methods to evaluate $\int^{10}_{1}f(x)$ where $f(x)=x^2+exp^{-x}+10/x$"""
"""
First, evaluate $\int^{10}_{1}f(x)$ analytically. In other words, integrate $f(x)$ by hand, and then evaluate the bounds in Python."""

#method 1: analytical solution
import numpy as np
def f(x):
    return (x**3)/3-np.exp(-x)+10*np.log(x)
print (f(10)-f(1), ' is the analytical solution')

"""Evaluate $\int^{10}_{1}f(x)$ using the trapezoidal rule using 5 evenly-spaced points (which yields 4 trapezoids).
"""
#method 2: Trapezoidal rule
def f1(x):
    return (x**2+np.exp(-x)+10/x)

N=5

dx=(10-1)/(N-1.)
#print ("The width of the trapezoid is "), dx
sum=0
x=1.
while x<10:
    sum=sum+2*f1(x)
    #print sum
    #print x
    x=x+dx
sum=sum+ f1(x)

print ("Integral by trapezoid method is", dx*sum*0.5)
```

```python
#Trapezoid rule with multiple N
import numpy as np
def f(x):
    return (x**3)/3-np.exp(-x)+10*np.log(x)
anal=f(10)-f(1)
print (anal, ' is the analytical solution')
Ymain=[]
Xmain=[]
def f1(x):
    return x**2+np.exp(-x)+10/x

for N in range(2,101):
    dx=(10-1)/(N-1.)
    #print ("The width of the trapezoid is "), dx
    sum=0
    x=1.
    while x<10:
        sum=sum+2*f1(x)
        #print sum
        #print x
        x=x+dx
    sum=sum+ f1(x)
    int=dx*sum*0.5
    err=abs(int-anal)
    print(err)
    Ymain.append(err)
    Xmain.append(N)
#print(Ymain)
#print(Xmain)

import matplotlib.pyplot as plt
plt.plot(Xmain, np.log(Ymain))
plt.xlabel('Number of points')
plt.ylabel('Error')
plt.show()
```
